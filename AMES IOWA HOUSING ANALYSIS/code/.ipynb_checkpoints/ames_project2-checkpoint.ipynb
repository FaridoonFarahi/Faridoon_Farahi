```python
# Since Lot Frontrage is Linear feet of street connected to property, hence NaN means that there is no Lot Frontage
# In other words, there is 0 feet of street connected to property
train['Lot Frontage'] = train['Lot Frontage'].fillna(0)
test['Lot Frontage'] = test['Lot Frontage'].fillna(0)
train.head()
```

```python
# Filter remaining columns with NaN values
nan_values = train[train.columns[train.isnull().any()]]
nan_values
```

```python
# Since the min value for columns except Mas Vnr Area,BsmtFin SF 1,BsmtFin SF 2,Bsmt Unf SF, Total Bsmt SF is 0, Garage Area 
# we cannot replace the NaN values in this column with 0. # Therefore we replace it with mean value.

# Calculating the mean values for the above mentioned columns
mean_MasVnrArea_train = train['Mas Vnr Area'].mean()
mean_bstfinsf1_train =  train['BsmtFin SF 1'].mean()
mean_bstfinsf2_train = train['BsmtFin SF 2'].mean()
mean_bstunfsf_train =  train['Bsmt Unf SF'].mean()
mean_totalbsmtsf_train =  train['Total Bsmt SF'].mean()
mean_garagearea_train =  train['Garage Area'].mean()

mean_MasVnrArea_test = test['Mas Vnr Area'].mean()
mean_bstfinsf1_test =  test['BsmtFin SF 1'].mean()
mean_bstfinsf2_test = test['BsmtFin SF 2'].mean()
mean_bstunfsf_test =  test['Bsmt Unf SF'].mean()
mean_totalbsmtsf_test =  test['Total Bsmt SF'].mean()
mean_garagearea_test =  test['Garage Area'].mean()

# Replacing NaN values with the mean values calculated above
train['Mas Vnr Area'].fillna(mean_MasVnrArea_train, inplace=True)
train['BsmtFin SF 1'].fillna(mean_bstfinsf1_train, inplace=True)
train['BsmtFin SF 2'].fillna(mean_bstfinsf2_train, inplace=True)
train['Bsmt Unf SF'].fillna(mean_bstunfsf_train, inplace=True)
train['Total Bsmt SF'].fillna(mean_totalbsmtsf_train, inplace=True)
train['Garage Area'].fillna(mean_garagearea_train, inplace=True)

test['Mas Vnr Area'].fillna(mean_MasVnrArea_test, inplace=True)
test['BsmtFin SF 1'].fillna(mean_bstfinsf1_test, inplace=True)
test['BsmtFin SF 2'].fillna(mean_bstfinsf2_test, inplace=True)
test['Bsmt Unf SF'].fillna(mean_bstunfsf_test, inplace=True)
test['Total Bsmt SF'].fillna(mean_totalbsmtsf_test, inplace=True)
test['Garage Area'].fillna(mean_garagearea_test, inplace=True)

# We replace the Bsmt Full Bath, bsmt Half Bath and Garage Cars with the min value since these cannot be replaced
# with mean value and also since customer is more concered with the min number of these features

# Calculating the min values for the above mentioned columns
min_bsmtfullbath_train = train['Bsmt Full Bath'].min()
min_bsmthalfbath_train = train['Bsmt Half Bath'].min()
min_garagecars_train = train['Garage Cars'].min()

min_bsmtfullbath_test = test['Bsmt Full Bath'].min()
min_bsmthalfbath_test = test['Bsmt Half Bath'].min()
min_garagecars_test = test['Garage Cars'].min()

# Replacing NaN values with the min values calculated above
train['Bsmt Full Bath'].fillna(min_bsmtfullbath_train, inplace=True)
train['Bsmt Half Bath'].fillna(min_bsmthalfbath_train, inplace=True)
train['Garage Cars'].fillna(min_garagecars_train, inplace=True)

test['Bsmt Full Bath'].fillna(min_bsmtfullbath_test, inplace=True)
test['Bsmt Half Bath'].fillna(min_bsmthalfbath_test, inplace=True)
test['Garage Cars'].fillna(min_garagecars_test, inplace=True)

# Since we cannot replace year with mean year, we replace Garage Yr Built NaN values w/ 0 indiciating the info isn't available
train['Garage Yr Blt'].fillna(0, inplace=True)
test['Garage Yr Blt'].fillna(0, inplace=True)
```

### Preprocessing and Modeling

### One hot label encoding

```python
# Use indexing to drop ID, PID and SalePrice columns from our train and create a new dataframe for one hot label encoding
train_ohe = train.iloc[:,2:-1]

# Get all the catergorical column names
cat_cols = (train_ohe.loc[:, train_ohe.dtypes == object].columns)

# Perform one hot label encoding
train_ohe = pd.get_dummies(train_ohe, columns=cat_cols)
display(train_ohe)
```

### Getting Ready to Fit Model

```python
# Get our X (features) and y is our target label
y = train['SalePrice']
X = train_ohe

# Split our data into train and validation set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
# Create the model and fit data (train it)
lr = LinearRegression()
lr.fit(X_train,y_train)

# Make predictions on the validation data (X_test)
y_pred_lr = lr.predict(X_test)

# Calculate Root Mean Squared Error
rmse_lr = mean_squared_error(y_test, y_pred_lr, squared=False)

# Calculate Mean Squared Error
mse_lr = mean_squared_error(y_test, y_pred_lr, squared=True)

# Calculate the Score (R^2)
score_lr = lr.score(X_train, y_train)

print('RMSE :', rmse_lr)
print('MSE  :', mse_lr)
print('R2   :',score_lr)
```

```python
rr = Ridge(alpha = 10)
rr.fit(X_train,y_train)

# Make predictions on the test_df
y_pred_rr = rr.predict(X_test)

# Calculate Root Mean Squared Error
rmse_rr = mean_squared_error(y_test, y_pred_rr, squared=False)

# Calculate Mean Squared Error
mse_rr = mean_squared_error(y_test, y_pred_rr, squared=True)

# Calculate the Score (R2)
score_rr = rr.score(X_train, y_train)

print('RMSE :', rmse_rr)
print('MSE  :', mse_rr)
print('R2   :',score_rr)
```

```python
llr = Lasso(alpha=0.1, normalize=True)
llr.fit(X_train,y_train)

# Make predictions on the test_df
y_pred_llr = llr.predict(X_test)

# Calculate Root Mean Squared Error
rmse_llr = mean_squared_error(y_test, y_pred_llr, squared=False)

# Calculate Mean Squared Error
mse_llr = mean_squared_error(y_test, y_pred_llr, squared=True)

# Calculate the Score (R2)
score_llr = llr.score(X_train, y_train)

print('RMSE :', rmse_llr)
print('MSE  :', mse_llr)
print('R2   :',score_llr)
```

### Cross Validation

```python
'''Perform Cross Validation with 5 folds'''

# Calculate scores with Linear Regression
lr_cv_scores = cross_val_score(lr, X_train, y_train, cv=5)
print('Linear Regression CV Mean :', lr_cv_scores.mean())

# Calculate scores with Ridge Regression
rr_cv_scores = cross_val_score(rr, X_train, y_train, cv=5)
print('Ridge Regression CV Mean  :', rr_cv_scores.mean())

# Calculate scores with Lasso Regression
llr_cv_scores = cross_val_score(llr, X_train, y_train, cv=5)
print('Lasso Regression CV Mean  :', llr_cv_scores.mean())
```
